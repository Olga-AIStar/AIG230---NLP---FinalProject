### Removing Duplicated Complaints

Based on the statistics, the [Consumer complaint narrative] column contains multiple identical entries (e.g., counts of 27,389, 23,655, etc.) that appear to be used as templates. Keeping these as they are will lead to significant problems when training NLP models:

- Data Leakage and Overfitting
    If these identical complaints appear in both training and test sets, the model won't actually "learn" how to classify language; it will simply memorize specific legal templates. This creates a false sense of high accuracy that will fail when the model encounters real, unique consumer stories.

- Over-representation of "Legalese"
    The most frequent record appears over 27,000 times. This specific legal jargon ("In accordance with the Fair Credit Reporting Act...") will dominate in vocabulary. The model may begin to assume that these specific legal citations are the only indicators of a credit reporting issue, causing it to ignore natural human descriptions of the same problem.

- Skewed Class Distribution
    These duplicates are almost exclusively related to Credit Reporting. If they are not removed, the "Credit Reporting" category becomes artificially inflated. This creates a heavily biased model that will likely default to guessing "Credit Reporting" for almost any input, simply because it was trained on so many identical copies of that class.

```python
initial_count = len(complains_with_concent)

df = complains_with_concent.drop_duplicates(subset=['Consumer complaint narrative']).copy()

print(f'Duplicated narratives removed: {initial_count - len(df):,}; current size is {len(df):,} records')
```

### Grouping 21 products  into 8 high-level categories.

```python
# 1. The mapping dictionary
product_mapping = {
    # Credit Reporting
    "Credit reporting or other personal consumer reports": "Credit Reporting",
    "Credit reporting, credit repair services, or other personal consumer reports": "Credit Reporting",
    "Credit reporting": "Credit Reporting",
    
    # Debt Collection
    "Debt collection": "Debt Collection",
    
    # Banking
    "Checking or savings account": "Banking & Deposits",
    "Bank account or service": "Banking & Deposits",
    
    # Cards
    "Credit card or prepaid card": "Cards",
    "Credit card": "Cards",
    "Prepaid card": "Cards",
    
    # Mortgage
    "Mortgage": "Mortgage",
    
    # Loans & Lending
    "Student loan": "Loans & Lending",
    "Vehicle loan or lease": "Loans & Lending",
    "Payday loan, title loan, or personal loan": "Loans & Lending",
    "Payday loan, title loan, personal loan, or advance loan": "Loans & Lending",
    "Consumer Loan": "Loans & Lending",
    "Payday loan": "Loans & Lending",
    
    # Money Services
    "Money transfer, virtual currency, or money service": "Money Services",
    "Money transfers": "Money Services",
    "Virtual currency": "Money Services",
    
    # Others
    "Debt or credit management": "Other Financial Services",
    "Other financial service": "Other Financial Services"
}


df['Product_Group'] = df['Product'].map(product_mapping)

# 3. Optional: Handle missing values
# If a value wasn't in the dictionary, it will become NaN. 
# You can fill those with 'Other' or keep the original name.
df['Product_Group'] = df['Product_Group'].fillna('Other')

# Verify the results
print(df['Product_Group'].value_counts().apply(lambda x: f"{x:,}"))
```

The following code is designed to address the challenges of working with a massive, imbalanced dataset. By implementing a stratified sampling approach, we ensure that our data is both manageable for training and representative of all categories.

Scaling Down for Training: 3.7 million records is too large to train on a local machine efficiently. Take a stratified random sample (e.g., 100,000 records) that maintains the proportional representation of product groups.

```python
sampling_column_name = "Product_Group"

target_sample_size = cfg.target_sample_size

if len(df) > target_sample_size:
    df_sampled, _ = train_test_split( df, train_size=target_sample_size, stratify=df[sampling_column_name], random_state=42)
else:
    df_sampled = df

df_sampled.shape
```

#### To ensure the model treats every product group with equal importance implemented a hybrid Under-sampling and Over-sampling strategy.
each group has fixed target_size (defined in config). 

**Under-sampling (Large Categories):** If a category has more than the target number of records, used replace=False to randomly select a subset. This trims down the "noisy" majority classes to a manageable size.

**Over-sampling (Small Categories):** If a category is too small, used replace=True to duplicate existing records until they reach the target. This ensures that rare but important categories (like "Student Loans") have enough presence for the model to learn their specific patterns.

```python
balanced_chunks = []
target_size = cfg.product_target_size

for product_name in df_sampled[sampling_column_name].unique():
    product_subset = df_sampled[df_sampled[sampling_column_name] == product_name] 

    if len(product_subset) > target_size:
        product_resample = resample(product_subset, replace=False, n_samples=target_size, random_state=42)
    else:
        product_resample = resample(product_subset, replace=True, n_samples=target_size, random_state=42)

    balanced_chunks.append(product_resample)

df_final = pd.concat(balanced_chunks)

print(f"Final dataset size: {len(df_final):,}")
print(df_final[sampling_column_name].value_counts().apply(lambda x: f"{x:,}"))

```
